In the field of data-related research, it is very important to handle missing data either by deleting or imputation(handling the missing values with some estimation).

Different Methods of Dealing With Missing Data
1. Deleting the column with missing data
2. Deleting the row with missing data
3. Filling the Missing Values – Imputation
   
   (i) Numerical data - use mean
   
   (ii) categorical data 
       - use mode, median
       - assign the NaN values their own category
   
   
4. Advanced Imputation

### Creating own datasets

```python
students = {
    "name": ['Rishabh', "Yogesh", 'rahul', "sanjay", 'rohit', 'ram','raja', 'manish', 'ravi','Mohit'],
    'marks': {
        "physics" : [82,86,np.nan,56,np.nan,49,64,92,np.nan,81],
        "maths" : [np.nan,70,75,76,83,np.nan,85,63,98,np.nan],
        'Chem':  [46, np.nan, 56,74,92,82,np.nan,54,64,np.nan]  
    }
}

students
```

```python
marks_df = pd.DataFrame(students['marks'])
marks_df
name_df = pd.DataFrame(students['name'])
name_df
```

```python
df = pd.concat([name_df,marks_df], axis = 1)
df
```

```python
# now we have data and there is need to detect missing values
num_rows = df.shape[0]
num_missing = num_rows - df.count()
print(num_missing)
df.count()          #  used to count the number of non-null 

# subsitiute
df.isnull().sum()
```

```python
# copying the data 
df1 = df.copy()
df1
```

```python
# now removing the row
df1.dropna()
```

```python
mean_physics = df['physics'].mean()
df1["physics"].fillna(mean_physics, inplace = True)
df1
```

```python
# now go for maths by using another way
mean_physics = df['maths'].mean()
df1["maths"] = df1["maths"].fillna(mean_physics)
df1
```

```python
mean = df1['Chem'].mean()
df1["Chem"].fillna(mean,inplace = True)
df1
```

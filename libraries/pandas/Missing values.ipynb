In the field of data-related research, it is very important to handle missing data either by deleting or imputation(handling the missing values with some estimation).

Different Methods of Dealing With Missing Data
1. Deleting the column with missing data
2. Deleting the row with missing data
3. Filling the Missing Values – Imputation
   
   (i) Numerical data - use mean
   
   (ii) categorical data 
       - use mode, median
       - assign the NaN values their own category
   
   
4. Advanced Imputation

#### Numeric data (Use Mean)

```python
mean = df3["reviews_per_month"].mean()
mean
df3["reviews_per_month"].fillna(mean)
df3
```

```python
df3.head()
df3.isna().sum()
```

```python
df3[df3["number_of_reviews"]==0].head(2)
```

```python
df3.isna().sum()
```

#### Categorical Data (Use mode value)

```python
df3["last_review"].unique()
    
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0])
```

#### Categorical data (Use new category)

```python
df3["last_review"].fillna("Not Reviewed", inplace = True)
```

```python
df3.isna().sum()
```

```python
df3["host_name"].fillna("Unknown")
```

```python
# Forward Fill (ffill()): Propagates the last valid observation forward to fill gaps.
# Uses the next valid observation to fill gaps (bfill())
df3["name"].ffill()
```

## dealing with numeric data in the dataset

```python
os.getcwd()
data = pd.read_csv('googleplaystore.csv').dropna()
(data)
```

```python
avg_rating = int(data["Rating"].sum()/9360)
print(f"Average rating of the apps is {avg_rating}")

# another way
data["Rating"].mean()  # returns the mean of the 'Rating' column
data["Rating"].median()  # returns the median of the 'Rating' column
data["Rating"].mode()  # returns the mode of the 'Rating' column
data["Rating"].std()  # returns the standard deviation of the 'Rating' column
data["Rating"].var()  # returns the variance of the 'Rating' column
data["Rating"].describe()  # returns the descriptive statistics of the 'Rating' column

```

```python
# checking the number of apps with rating 5.0
s = 0
for i in data["Rating"]:
    if i == 5.0:
        s += 1
print(f"Number of apps with rating 5.0 is {s}")
```

```python
data['Reviews'] = data['Reviews'].astype(int)
data.dtypes

```

```python
avg_review = data ["Reviews"].sum()
count = data["Reviews"].count()
print(f"Average review of the apps is {avg_review/count}")  
```

```python
cat = data["Category"].unique()  # returns the unique values in the 'Category' column
count = 0
for i in cat :
    count += 1
    print(i)
print("The total no of categories are",count)

print("The total no of categories are",len(cat))
 
```

```python
# counting the specfic categories in the dataset
count  = 0
for i in data["Category"]:
    if i == "ART_AND_DESIGN":
        count += 1
print("The total no of ART_AND_DESIGN categories are",count)
```

```python
type_ = data["Type"]
free = 0
paid = 0
for i in type_:
    if i == "Free":
        free += 1
    else:
        paid += 1
print("total free values are",free)
print("total paid values are ",paid)
    
sum = (free+paid)
ratio = free/paid
print(ratio)
# free percentage
free_percentage = (free/sum)*100
print(free_percentage)
```

```python
for name in data["Category"].unique():
    count  = 0
    for i in data["Category"]:
        if i == name:
            count += 1
    print(name ,':',count)
```

```python
# converting the above data into dictionary format
category_count = {}
for name in data["Category"].unique():
    count  = 0
    for i in data["Category"]:
        if i == name:
            count += 1
    category_count[name]=count
```

```python
import numpy as np
import pandas as pd
```

```python
lst  = [1, 2, 3, 4, 5]
array = np.array(lst)
print(array)
print(type(array))
print("*"*25)

# Convert the list to a pandas Series
series = pd.Series(lst)
print(series)
print(type(series))
print("*"*25)

# Convert the list to a pandas DataFrame
dataframe = pd.DataFrame(lst, columns=["Numbers"])
print(dataframe)
print(type(dataframe))
```

## Selection

```python
data["Reviews"]  # select the "Reviews" column - getitem
```

```python
data[1:3]
```

```python
# selecting a row matching a label
print(data.loc[0] ) # get the row with index 1
print(data.head() ) # display the first few rows of the DataFrame

print("*"*25)


# selecting all rows (:) with a select columns labels
data.loc[:, ["App", "Category", "Rating"]]  # select all rows and the "App", "Category", and "Rating" columnsf
```

```python
# For label slicing, both endpoints are included:
data.loc[0:2, ["App", "Category", "Rating"]]  # select rows 0 to 2 and the "App", "Category", and "Rating" columns
```

```python
# selecting a single row and column label returns a scalar
data.loc[[0], "App"]  # select the "App" column of the first row
data.loc[0, "App"]
```

### Selection by Position

```python
# select via the position of the passed integers
data.iloc[4]

# intger slicing acts similar to Numpy/Python
data.iloc[3:5,0:2]
```

```python
# list of intger position locations
data.iloc[[0, 2, 4], [0, 1, 2]]  # select rows 0, 2, and 4 and columns 0, 1, and 2
```

```python
# for slicing columns explicitly
data.iloc[:, 0:2]  # select all rows and columns 0 and 1

# for slicing rows explicitly
data.iloc[0:2, :]
```

```python
# for getting a value explicitly
data.iloc[1,1]
# For getting fast access to a scalar (equivalent to the prior method):
data.iat[1, 1]  # get the value at row 1 and column 1
```

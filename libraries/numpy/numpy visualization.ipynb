```python
x = np.linspace(0,10,1000)
# print(x)
y_ = np.sin(x)
y = np.cos(x)
# print(y)
# plotting
plt.plot(x,y,'o',linestyle="--",markersize = 1,alpha = 0.5)
plt.plot(x,y_,'o',linestyle="--",markersize = 1,alpha = 0.5,color = "blue")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sine and Cosine Functions")
plt.legend(["cos(x)","sin(x)"])
plt.grid()
```

```python
import numpy as np  # Importing NumPy library for numerical operations
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating a 10x5 matrix with random values between 0 and 1
data = np.random.rand(10, 5)
print(data[0])
print(data[1])
print(data[2])
print(data[3])
print(data[4])
# Plotting stacked bar charts for each dataset by adding columns progressively
# The bars represent the sum of columns in data, starting with the first column (D-1) and adding subsequent columns for each subsequent dataset (D-2 to D-5)
plt.bar(np.arange(10), data[:, 0] + data[:, 1] + data[:, 2] + data[:, 3] + data[:, 4], label='D-5')
plt.bar(np.arange(10), data[:, 0] + data[:, 1] + data[:, 2] + data[:, 3], label='D-4')
plt.bar(np.arange(10), data[:, 0] + data[:, 1] + data[:, 2], label='D-3')
plt.bar(np.arange(10), data[:, 0] + data[:, 1], label='D-2')
plt.bar(np.arange(10), data[:, 0], label='D-1')

# Adding a legend to label each dataset (D-1 to D-5)
plt.legend()

# Displaying the stacked bar chart
plt.show()
```

```python
# group bar graph
import numpy as np  # Importing NumPy library for numerical operations
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Creating a 3x3 matrix (3 rows and 3 columns) with some data
data = np.array([[10, 20, 30],
                 [40, 50, 60],
                 [70, 80, 90]])

# Setting the width of the bars in the bar chart
width = 0.25

# Creating an array for the x positions of the bars
x = np.arange(3)

# Plotting the first set of bars for D-1 (shifted to the left by 'width')
plt.bar(x - width, data[:, 0], width, label='D-1')

# Plotting the second set of bars for D-2 (centered on x positions)
plt.bar(x, data[:, 1], width, label='D-2')

# Plotting the third set of bars for D-3 (shifted to the right by 'width')
plt.bar(x + width, data[:, 2], width, label='D-3')

# Adding the legend to label each data set
plt.legend()

# Displaying the bar chart
plt.show()
```

```python
# histogram plot
data  = np.random.randn(10000000)  # Generating 1000 random numbers from a normal distribution
plt.hist(data, bins=1000, alpha=0.5, color='blue')  # Creating a histogram with 30 bins
plt.xlabel('Value')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis 
plt.title('Histogram of Random Data')  # Title of the histogram
plt.grid()  # Adding a grid to the plot
plt.show()  # Displaying the histogram
```

```python
# accumulative histogram plot
data  = np.random.randn(1000)  # Generating 1000 random numbers from a normal distribution
plt.hist(data, bins=10, alpha=0.5, color='blue', cumulative=True)  # Creating a histogram with 30 bins
plt.xlabel('Value')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis 
plt.title('Histogram of Random Data')  # Title of the histogram
plt.grid()  # Adding a grid to the plot
plt.show()  # Displaying the histogram
```

```python
# histogram plot with multiple datasets
data1 = np.random.randn(100000)  # Generating 1000 random numbers from a normal distribution
data2 = np.random.randn(100000) +6 # Generating another 1000 random numbers from a normal distribution
plt.hist(data1, bins=100, alpha=0.5, color='blue', label='Dataset 1')  # Creating a histogram for the first dataset
plt.hist(data2, bins=100, alpha=0.5, color='red', label='Dataset 2')  # Creating a histogram for the second dataset
plt.xlabel('Value')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis
plt.title('Histogram of Random Data')  # Title of the histogram
plt.legend()  # Adding a legend to label each dataset
plt.grid()  # Adding a grid to the plot
plt.show()  # Displaying the histogram
```

```python
# piechart
import numpy as np  # Importing NumPy library for numerical operations
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Defining the labels for the pie chart (programming languages)
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']

# Defining the corresponding popularity sizes for each language
sizes = [21.3, 14.6, 9.2, 8.5, 7.7]

# Defining the explode values for each slice (to offset certain slices for emphasis)
explode = (0.13, 0.0, 0, 0, 0)

# Defining custom colors for each slice of the pie
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']

# Plotting the pie chart with labels, colors, and percentages displayed
plt.title("Programming Language Popularity")  # Title for the pie chart
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.0f%%', shadow=True, explode=explode, startangle=140)

plt.axis('equal')  # Ensuring the pie chart is drawn as a circle (equal aspect ratio)
plt.legend(title="Languages", loc="upper left")  # Adding a legend to label each programming language

plt.show()  # Displaying the pie chart
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
```
